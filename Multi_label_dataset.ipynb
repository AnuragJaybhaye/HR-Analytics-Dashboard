```python
import torch
from torch.utils.data import TensorDataset, DataLoader
from datasets import load_dataset
from sentence_transformers import SentenceTransformer
```

```python
# 1. Setup
DEVICE = "cuda"
BATCH_SIZE = 50
sbert = SentenceTransformer('all-mpnet-base-v2')
NUM_LABELS = 28
```

```python
dataset = load_dataset("go_emotions", "simplified")
train_data = dataset["train"]
texts = train_data["text"]
```

```python
embeddings = sbert.encode(texts, convert_to_tensor=True)
```

```python
import torch.nn.functional as F
embeddings = F.normalize(embeddings, p=2, dim=1)
# normalize embeeding
```

```python
def multi_hot_encode(label_indices, num_classes):
    target = torch.zeros(num_classes)
    target[label_indices] = 1.0
    return target
```

```python
labels = torch.stack([multi_hot_encode(l, NUM_LABELS) for l in train_data["labels"]])
```

```python
labels[107]
```

```python
embeddings.shape
```

```python
train_dataset = TensorDataset(embeddings.cpu(), labels)
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
```

```python
#prepare test loader
test_data = dataset["test"]
texts = test_data["text"]
embeddings = sbert.encode(texts, convert_to_tensor=True, show_progress_bar=True)
embeddings = F.normalize(embeddings, p=2, dim=1)
labels = torch.stack([multi_hot_encode(l, NUM_LABELS) for l in test_data["labels"]])
test_dataset = TensorDataset(embeddings.cpu(), labels)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=True)
```

```python
import torch.nn as nn


class NeuralNetwork(nn.Module):

  def __init__(self):

    super(NeuralNetwork,self).__init__()
    self.layer1=nn.Linear(768,10000)
    self.layer2=nn.Linear(10000,5000)
    self.layer3=nn.Linear(5000,2000)
    self.layer4=nn.Linear(2000,1000)
    self.layer5=nn.Linear(1000,500)
    self.layer6 = nn.Linear(500,28)

  def forward(self,sentence_embeddings):

    out=torch.relu(self.layer1(sentence_embeddings))
    out=torch.relu(self.layer2(out))
    out=torch.relu(self.layer3(out))
    out=torch.relu(self.layer4(out))
    out = torch.relu(self.layer5(out))
    out=self.layer6(out)

    return out
```

```python
model = NeuralNetwork()
```

```python
for batch,label in train_loader:
  output = model(batch)
  print(output.shape)
  break
# this should return 1,28
```

```python
# Define the Loss function

loss = nn.BCEWithLogitsLoss()
```

```python
# Define the optimiser

optimiser = torch.optim.Adam(model.parameters(), lr=0.0001)
```

```python
model.to(DEVICE)
```

```python
epochs = 10

for epoch in range(epochs):
    current_loss = 0
    num_batches = 0
    for batch,label in train_loader:
        num_batches += 1
        batch = batch.to(DEVICE)
        label = label.to(DEVICE)
        output = model(batch)
        loss_value = loss(output, label)
        current_loss += loss_value
        loss_value.backward()
        optimiser.step()

    print(f"epoch {epoch} loss : {current_loss/num_batches}")
```

```python
sentence = "i am very happy today"
embedding = sbert.encode(sentence, convert_to_tensor=True)
embedding = F.normalize(embedding, p=2, dim=0)
model(embedding)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

all_preds, all_labels = [], []

with torch.no_grad():
    for data, labels in train_loader:
        data, labels = data.to(DEVICE), labels.to(DEVICE)
        outputs = model(data);
        all_preds.append(torch.sigmoid(outputs).cpu())
        all_labels.append(labels.cpu())
```

```python
from sklearn.metrics import f1_score
import numpy as np

preds = torch.cat(all_preds).numpy()
labels = torch.cat(all_labels).numpy()
optimal_thresholds = np.zeros(28)

# Removed values_to_test as we use all thresholds from roc_curve

for c in range(28):
    try:
        class_labels = labels[:, c]
        class_preds = preds[:, c]

        # Calculate ROC curve data points: FPR, TPR, and all thresholds
        fpr, tpr, thresholds = roc_curve(class_labels, class_preds)

        # Calculate Youden's J Statistic: J = TPR - FPR
        J = tpr - fpr

        # Find the threshold that maximizes J
        best_threshold = thresholds[np.argmax(J)]

        # Store and report AUROC (Area Under ROC Curve)
        optimal_thresholds[c] = best_threshold
        auc = roc_auc_score(class_labels, class_preds)
        print(f"Class {c}: Threshold={best_threshold:.3f}, Max J={np.max(J):.4f}, AUROC={auc:.4f}")

    except ValueError:
        optimal_thresholds[c] = 0.5 # Default if class is uniform (roc_curve raises ValueError)
        print(f"Class {c}: Uniform class, default Threshold=0.5")

print(f"\nOptimal thresholds array found: {optimal_thresholds}")
```

```python
def predict(sentence):
  prediction = torch.zeros(NUM_LABELS)
  embedding = sbert.encode(sentence, convert_to_tensor=True)
  # If embedding is a 1D tensor (single sentence), unsqueeze to make it 2D (batch_size=1)
  if embedding.dim() == 1:
    embedding = embedding.unsqueeze(0)
  embedding = F.normalize(embedding, p=2, dim=1) # Normalize along dimension 1 (features)
  output = model(embedding.to(DEVICE)) # Ensure input is on the correct device
  output = torch.sigmoid(output)
  for i in range(NUM_LABELS):
    # Access the output of the first (and only) item in the batch. output[0] is now a 1D tensor of probabilities.
    if output[0][i].item() > optimal_thresholds[i]:
      prediction[i] = 1
  return prediction
```
